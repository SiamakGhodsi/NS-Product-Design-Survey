```python
def is_valid_venue(venue_str):
    """Check if the venue string matches any of our target venue variants."""
    if venue_str:
        lower_venue = venue_str.lower()
        for variants in VENUE_VARIANTS.values():
            for variant in variants:
                if variant.lower() in lower_venue:
                    return True
    return False

def search_dblp(query, start=0, hits=100):
    """
    Query the DBLP API for publications matching the query.
    'hits' limits the number of results per page.
    'start' sets the offset for pagination.
    """
    url = "https://dblp.org/search/publ/api"
    params = {"q": query, "format": "json", "h": hits, "f": start}
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        return data.get("result", {}).get("hits", {}).get("hit", [])
    return []

    # --- DBLP Functions (with pagination) ---



def fetch_all_results(query, hits=100, max_pages=10):
    """
    Retrieve results from DBLP with pagination.
    """
    all_results = []
    start = 0
    for _ in range(max_pages):
        results = search_dblp(query, start, hits)
        if not results:
            break
        all_results.extend(results)
        start += hits
        #time.sleep(0.5)  # be respectful to the server
        time.sleep(2)  # be respectful to the server
    return all_results

def get_bibtex_from_dblp(entry):
    """Retrieve the BibTeX citation from a DBLP entry by appending '.bib' to its URL."""
    dblp_url = entry.get("info", {}).get("url", "")
    if dblp_url:
        bib_url = dblp_url + ".bib"
        response = requests.get(bib_url)
        if response.status_code == 200:
            return response.text
    return None
```

# Policy

## Search based on topic queries

### Topic Queries:
The topic_queries dictionary defines five search topics (each with a specific query string). These queries are designed to capture papers on different aspects of multi-objective reinforcement learning.

### Pagination & Filtering:
The helper functions search_dblp and fetch_all_results allow pagination. In process_topic(), each paper is filtered by publication year (≥1980) and whether its venue is in our list.

### Saving Results:
For each topic, the code saves a CSV file with the papers, writes all BibTeX citations to a topic-specific .bib file, and generates a bar chart (saved as an SVG file) showing the number of papers per year.

### Summary Chart:
Finally, a summary chart is generated that shows the total number of papers found per topic.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import time



def process_topic(topic, query):
    """
    For a given topic (with its query string), fetch DBLP results,
    filter by publication year and valid venue, and return a dictionary
    of papers.
    """
    papers = {}
    print(f"---\nSearching DBLP for topic: {topic}\nQuery: {query}")
    results = fetch_all_results(query, hits=100, max_pages=5)
    print(f"Fetched {len(results)} raw results.")
    for entry in results:
        info = entry.get("info", {})
        title = info.get("title", "").strip()
        year_str = info.get("year", "").strip()
        try:
            year = int(year_str)
        except:
            continue
        if year < YEAR_THRESHOLD:
            continue
        venue_raw = info.get("venue", "")
        if isinstance(venue_raw, list):
            venue = " ".join(venue_raw).strip()
        else:
            venue = venue_raw.strip()
        # Skip if venue is not in our combined list
        if not (title and venue and is_valid_venue(venue)):
            continue
        # Accept the paper if it meets our criteria
        if title not in papers:
            bibtex = get_bibtex_from_dblp(entry)
            papers[title] = {
                "title": title,
                "year": year,
                "venue": venue,
                "bibtex": bibtex,
                "source": "DBLP",
                "topic": topic
            }
    print(f"Topic '{topic}': {len(papers)} papers after filtering.")
    return papers

def plot_year_distribution(df, topic, filename):
    """
    Plot and save a bar chart for the distribution of papers per year for a given topic.
    """
    # Filter to papers from the given topic
    df_topic = df[df['topic'] == topic]
    df_topic['year'] = pd.to_numeric(df_topic['year'], errors='coerce')
    df_topic = df_topic[df_topic['year'] >= YEAR_THRESHOLD]
    if df_topic.empty:
        print(f"No data to plot for topic {topic}.")
        return
    year_counts = df_topic['year'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    plt.bar(year_counts.index.astype(int).astype(str), year_counts.values)
    plt.xlabel("Year")
    plt.ylabel("Number of Papers")
    plt.title(f"Frequency of Papers per Year (>=1980) for {topic}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved chart for '{topic}' as {filename}")
    plt.close()

def plot_summary_chart(topic_counts, filename):
    """
    Plot and save a summary chart showing the total paper count per topic.
    """
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]
    plt.figure(figsize=(10, 6))
    plt.bar(topics, counts)
    plt.xlabel("Topic")
    plt.ylabel("Total Number of Papers")
    plt.title("Total Papers per Topic")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved summary chart as {filename}")
    plt.close()

def main():
    # Dictionary to store papers for each topic
    all_papers = {}
    # To hold total counts per topic for summary chart
    topic_counts = {}

    # Process each topic
    for topic, query in topic_queries.items():
        papers = process_topic(topic, query)
        all_papers.update(papers)
        topic_counts[topic] = len(papers)
        # Save CSV for the topic
        if papers:
            df_topic = pd.DataFrame(list(papers.values()))
            csv_filename = f"{topic.replace(' ', '_')}_papers.csv"
            df_topic.to_csv(csv_filename, index=False)
            print(f"Saved CSV for topic '{topic}' as {csv_filename}")
            # Save BibTeX citations for the topic
            bib_filename = f"scraped_bibtex_{topic.replace(' ', '_')}.bib"
            with open(bib_filename, "w", encoding="utf-8") as f:
                for paper in papers.values():
                    if paper["bibtex"]:
                        f.write(paper["bibtex"] + "\n\n")
            print(f"Saved BibTeX for topic '{topic}' as {bib_filename}")
            # Generate chart for the topic
            plot_filename = f"{topic.replace(' ', '_')}_by_year.svg"
            plot_year_distribution(df_topic, topic, plot_filename)

    # Save a combined CSV for all topics
    if all_papers:
        df_all = pd.DataFrame(list(all_papers.values()))
        df_all.to_csv("all_topics_papers.csv", index=False)
        print("Saved combined CSV for all topics as all_topics_papers.csv")

    # Generate summary chart for all topics
    plot_summary_chart(topic_counts, "summary_topic_counts.svg")

if __name__ == "__main__":
    main()
```

- Neuro‐symbolic: variations like "neuro symbolic", "neuro‑symbolic", etc.
- Multi‐objective optimization: variations like "multi-objective", "multi objective", "Pareto", etc.

## Generates Queries Systematically:
It creates all unique combinations for two‐set groups (RL+NS, RL+MO, NS+MO) and a three‑set group (RL+NS+MO).
(You can adjust the number of pages or hits per query if needed.)

## Scrapes DBLP:
For each generated query it fetches results (using pagination), filters them (by publication year ≥1980 and valid venue), and aggregates unique papers per topic group.

## Saves the Data:
For each topic group it saves CSV files and concatenates BibTeX entries into a .bib file.

## Generates Visualizations Inline:
It creates per‐topic bar charts (using a different color per topic) showing the paper count per year, and a final summary chart of total counts per topic.
All charts are saved as SVG files and shown inline.

```python
import os
import time
import itertools
import requests
import pandas as pd
import matplotlib.pyplot as plt

# --- Set working directory (adjust as needed) ---
print("Current directory:", os.getcwd())
os.chdir("/content/drive/My Drive/Colab Notebooks/Survey Paper/Neuro-Symbolic Multi-objective RL/")
print("New directory:", os.getcwd())

OUTPUT_DIR = "bipartite"
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)


# Generate query combinations (without repetition) for different groups:
import itertools

# For RL + NS (include both orders)
comb_rl_ns = list(itertools.product(rl_keywords, ns_keywords))
queries_rl_ns = [" ".join(pair) for pair in comb_rl_ns]
comb_ns_rl = list(itertools.product(ns_keywords, rl_keywords))
queries_ns_rl = [" ".join(pair) for pair in comb_ns_rl]
queries_rl_ns_total = list(set(queries_rl_ns + queries_ns_rl))

# For RL + MO (include both orders)
comb_rl_mo = list(itertools.product(rl_keywords, mo_keywords))
queries_rl_mo = [" ".join(pair) for pair in comb_rl_mo]
comb_mo_rl = list(itertools.product(mo_keywords, rl_keywords))
queries_mo_rl = [" ".join(pair) for pair in comb_mo_rl]
queries_rl_mo_total = list(set(queries_rl_mo + queries_mo_rl))

# For NS + MO (include both orders)
comb_ns_mo = list(itertools.product(ns_keywords, mo_keywords))
queries_ns_mo = [" ".join(pair) for pair in comb_ns_mo]
comb_mo_ns = list(itertools.product(mo_keywords, ns_keywords))
queries_mo_ns = [" ".join(pair) for pair in comb_mo_ns]
queries_ns_mo_total = list(set(queries_ns_mo + queries_mo_ns))

# For RL + NS + MO: Generate all possible permutations for each combination
queries_rl_ns_mo_total = set()
for r in rl_keywords:
    for n in ns_keywords:
        for m in mo_keywords:
            for perm in itertools.permutations([r, n, m]):
                queries_rl_ns_mo_total.add(" ".join(perm))
queries_rl_ns_mo_total = list(queries_rl_ns_mo_total)

# Combine all query groups into a dictionary:
combination_queries = {
    "RL+NS": queries_rl_ns_total,
    "RL+MO": queries_rl_mo_total,
    "NS+MO": queries_ns_mo_total,
    "RL+NS+MO": queries_rl_ns_mo_total
}


# --- Processing Each Topic Group ---

def process_topic_group(topic, query_list):
    """
    For a given topic group and its list of query strings,
    fetch DBLP results, filter by publication year (>= YEAR_THRESHOLD), valid venue,
    and ensure that the combined text from title and abstract contains all tokens from the query.
    Returns a dictionary of unique papers.
    """
    papers = {}
    print(f"Processing topic group: {topic} ...")
    for query in query_list:
        print(f"  Query: {query}")
        results = fetch_all_results(query=query, hits=100, max_pages=3)
        print(f"    Fetched {len(results)} raw results.")
        for entry in results:
            info = entry.get("info", {})
            title = info.get("title", "").strip()
            year_str = info.get("year", "").strip()
            try:
                year = int(year_str)
            except:
                continue
            if year < YEAR_THRESHOLD:
                continue
            venue_raw = info.get("venue", "")
            if isinstance(venue_raw, list):
                venue = " ".join(venue_raw).strip()
            else:
                venue = venue_raw.strip()
            # Get abstract (if available) and combine with title
            abstract = info.get("abstract", "")
            combined_text = (title + " " + abstract).lower()
            # Check that every token in the query is present in the combined text.
            if not all(token in combined_text for token in query.lower().split()):
                continue
            if not (title and venue and is_valid_venue(venue)):
                continue
            if title not in papers:
                bibtex = get_bibtex_from_dblp(entry)
                papers[title] = {
                    "title": title,
                    "year": year,
                    "venue": venue,
                    "bibtex": bibtex,
                    "source": "DBLP",
                    "topic": topic
                }
    print(f"Topic group '{topic}': {len(papers)} papers after filtering.\n")
    return papers

# --- Plotting Functions ---

def plot_year_distribution(df, topic, filename, color):
    """
    Plot and save a bar chart for the distribution of papers per year for a given topic.
    """
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    df_topic = df[df['year'] >= YEAR_THRESHOLD]
    if df_topic.empty:
        print(f"No data to plot for topic {topic}.")
        return
    year_counts = df_topic['year'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    plt.bar(year_counts.index.astype(int).astype(str), year_counts.values, color=color)
    plt.xlabel("Year")
    plt.ylabel("Number of Papers")
    plt.title(f"Papers per Year (>=1980) for {topic}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved chart for '{topic}' as {filename}")
    plt.show()  # show inline

def plot_summary_chart(topic_counts, filename):
    """
    Plot and save a summary chart showing the total paper count per topic group.
    """
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]
    colors = ["blue", "green", "red", "purple"]
    plt.figure(figsize=(10, 6))
    plt.bar(topics, counts, color=colors[:len(topics)])
    plt.xlabel("Topic Group")
    plt.ylabel("Total Number of Papers")
    plt.title("Total Papers per Topic Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved summary chart as {filename}")
    plt.show()

# --- Main Execution ---

def main():
    all_papers = {}
    topic_counts = {}

    # Define a color map for each topic group for visualization
    color_map = {
        "RL+NS": "blue",
        "RL+MO": "green",
        "NS+MO": "red",
        "RL+NS+MO": "purple"
    }

    # Process each topic group from our combination queries dictionary
    for topic, queries in combination_queries.items():
        papers = process_topic_group(topic, queries)
        all_papers.update(papers)
        topic_counts[topic] = len(papers)

        # Save CSV for the topic group
        if papers:
            df_topic = pd.DataFrame(list(papers.values()))
            csv_filename = os.path.join(OUTPUT_DIR, f"{topic.replace('+', '_')}_papers.csv")
            df_topic.to_csv(csv_filename, index=False)
            print(f"Saved CSV for topic group '{topic}' as {csv_filename}")

            # Save BibTeX citations for the topic group
            bib_filename = os.path.join(OUTPUT_DIR, f"scraped_bibtex_{topic.replace('+', '_')}.bib")
            with open(bib_filename, "w", encoding="utf-8") as f:
                for paper in papers.values():
                    if paper["bibtex"]:
                        f.write(paper["bibtex"] + "\n\n")
            print(f"Saved BibTeX for topic group '{topic}' as {bib_filename}")

            # Generate and show year distribution chart
            plot_filename = os.path.join(OUTPUT_DIR, f"{topic.replace('+', '_')}_by_year.svg")
            plot_year_distribution(df_topic, topic, plot_filename, color_map.get(topic, "black"))

    # Save a combined CSV for all topic groups
    if all_papers:
        df_all = pd.DataFrame(list(all_papers.values()))
        combined_csv = os.path.join(OUTPUT_DIR, "all_topics_papers.csv")
        df_all.to_csv(combined_csv, index=False)
        print("Saved combined CSV for all topic groups as all_topics_papers.csv")

    # Generate summary chart for all topic groups
    plot_summary_chart(topic_counts, os.path.join(OUTPUT_DIR, "summary_topic_counts.svg"))


if __name__ == "__main__":
    main()
```

# Policy
### At least 50% combined keywords from different categories.

#### How It Works
Determining Required Categories:
The function checks the topic string (e.g., if it contains "RL", "NS", or "MO") and adds the corresponding global keyword lists to required_categories.

#### Overall Token Matching:
It splits the current query into tokens and counts how many appear in the combined title + abstract text. It then requires that at least 50% of the tokens (or a minimum of 2 tokens) are found.

#### Category Matching:
For each required category (for instance, for a topic like "RL+NS+MO" it requires tokens from all three groups), it checks that at least one token from that group's keyword list is present in the combined text.

```python
def process_topic_group(topic, query_list):
    """
    For a given topic group and its list of query strings,
    fetch DBLP results, filter by publication year (>= YEAR_THRESHOLD) and valid venue,
    and check that:
      - At least 50% of the tokens (min 2) from the query appear in the combined title+abstract, and
      - For each required category (determined by the topic key: "RL", "NS", "MO"),
        at least one token from that category appears in the combined text.
    Returns a dictionary of unique papers.
    """
    papers = {}
    print(f"Processing topic group: {topic} ...")

    # Determine required categories from the topic key:
    required_categories = []
    if "RL" in topic:
        required_categories.append(rl_keywords)
    if "NS" in topic:
        required_categories.append(ns_keywords)
    if "MO" in topic:
        required_categories.append(mo_keywords)

    for query in query_list:
        print(f"  Query: {query}")
        results = fetch_all_results(query, hits=100, max_pages=10)
        print(f"    Fetched {len(results)} raw results.")
        for entry in results:
            info = entry.get("info", {})
            title = info.get("title", "").strip()
            year_str = info.get("year", "").strip()
            try:
                year = int(year_str)
            except:
                continue
            if year < YEAR_THRESHOLD:
                continue
            venue_raw = info.get("venue", "")
            if isinstance(venue_raw, list):
                venue = " ".join(venue_raw).strip()
            else:
                venue = venue_raw.strip()
            # Get abstract (if available) and combine with title
            abstract = info.get("abstract", "")
            combined_text = (title + " " + abstract).lower()

            # Overall token matching: require at least 50% (min 2) of query tokens to appear.
            tokens = query.lower().split()
            token_count = sum(1 for token in tokens if token in combined_text)
            overall_threshold = max(2, len(tokens) / 2)
            if token_count < overall_threshold:
                continue

            # Category-based matching: require at least one token from each required category.
            category_pass = True
            for category in required_categories:
                if not any(token in combined_text for token in category):
                    category_pass = False
                    break
            if not category_pass:
                continue

            if not (title and venue and is_valid_venue(venue)):
                continue

            if title not in papers:
                bibtex = get_bibtex_from_dblp(entry)
                papers[title] = {
                    "title": title,
                    "year": year,
                    "venue": venue,
                    "bibtex": bibtex,
                    "source": "DBLP",
                    "topic": topic
                }
    print(f"Topic group '{topic}': {len(papers)} papers after filtering.\n")
    return papers

if __name__ == "__main__":
    main()
```

# Semantci Scholar Trick
The combination of NS + MO + RL of the previous keyword and their combinations with permutation non-repitetive but with order sensitivity, takes
- For NS + RL : $(6*7) + (7*6) = 6*7*2 = 84$
- For NS + MO : $(7*22) + (22*7) = 7*22*2 = 308$
- For RL + MO : $(6*22) + (22*6) = 6*22*2 = 264$

  then for the combination NS + MO + RL we have
- For NS + MO + RL : $(6*7*22) * 6 = 5544$   

Thus to limit the search space and since Smenatic Scholar API has its own NLP backend that doesn't need permutated queries so we don't permute and also reduce the combinations in the categories to generate the final combination NS+RL+MO

```python
import os
import time
import itertools
import requests
import pandas as pd
import matplotlib.pyplot as plt

# --- Set working directory ---
print("Current directory:", os.getcwd())
os.chdir("/content/drive/My Drive/Colab Notebooks/Survey Paper/Neuro-Symbolic Multi-objective RL/")
print("New directory:", os.getcwd())

# --- Create output subfolder "bipartite" if it doesn't exist ---
OUTPUT_DIR = "bipartite"
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
    print(f"Created subfolder: {OUTPUT_DIR}")


def is_valid_venue(venue_str):
    if venue_str:
        lower_venue = venue_str.lower()
        for variants in VENUE_VARIANTS.values():
            for variant in variants:
                if variant.lower() in lower_venue:
                    return True
    return False


# --- Semantic Scholar API Functions ---
def search_semantic_scholar(query, offset=0, limit=100, max_retries=5):
    url = "https://api.semanticscholar.org/graph/v1/paper/search"
    params = {
        "query": query,
        "offset": offset,
        "limit": limit,
        "fields": "title,year,venue,abstract,url"  # Removed doi field
    }
    retries = 0
    while retries < max_retries:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            try:
                data = response.json()
                return data.get("data", [])
            except Exception as e:
                print(f"Error parsing JSON for query: {query}, error: {e}")
                return []
        elif response.status_code == 429:
            print(f"429 Too Many Requests for query: {query}. Retrying after delay...")
            time.sleep(5)
            retries += 1
        else:
            print(f"Semantic Scholar API error {response.status_code} for query: {query}. Response: {response.text}")
            return []
    return []


def fetch_all_results_semantic(query, limit=100, max_pages=3):
    all_results = []
    offset = 0
    for _ in range(max_pages):
        results = search_semantic_scholar(query, offset, limit)
        if not results:
            break
        all_results.extend(results)
        offset += limit
        #time.sleep(1)
    return all_results

def get_citation_from_semantic(entry):
    title = entry.get("title", "")
    year = entry.get("year", "")
    venue = entry.get("venue", "")
    url = entry.get("url", "")
    citation = f"Title: {title}\nYear: {year}\nVenue: {venue}\nURL: {url}"
    return citation


# --- Processing Function for Semantic Scholar ---
def process_topic_group_semantic(topic, query_list):
    """
    For a given topic group and its list of query strings,
    fetch Semantic Scholar results, filter by publication year (>= YEAR_THRESHOLD)
    and valid venue, and ensure that for each required category (RL, NS, MO)
    at least one token from that category appears in the title or abstract.
    Returns a dictionary of unique papers.
    """
    papers = {}
    print(f"Processing topic group: {topic} ...")

    # Determine required categories based on the topic key.
    required_categories = []
    if "RL" in topic:
        required_categories.append(rl_keywords)
    if "NS" in topic:
        required_categories.append(ns_keywords)
    if "MO" in topic:
        required_categories.append(mo_keywords)

    for query in query_list:
        print(f"  Query: {query}")
        results = fetch_all_results_semantic(query, limit=100, max_pages=3)
        print(f"    Fetched {len(results)} raw results.")
        for entry in results:
            title = entry.get("title", "").strip()
            try:
                year = int(entry.get("year", 0))
            except:
                continue
            if year < YEAR_THRESHOLD:
                continue
            venue = entry.get("venue", "")
            if isinstance(venue, list):
                venue = " ".join(venue).strip()
            else:
                venue = venue.strip()
            abstract = entry.get("abstract", "")
            title_text = title.lower()
            abstract_text = abstract.lower() if abstract else ""
            combined_text = f"{title_text} {abstract_text}"

            # Check that each required category has at least one token in title or abstract.
            category_match = all(
                any(token in title_text or token in abstract_text for token in category)
                for category in required_categories
            )
            if not category_match:
                continue
            if not (title and venue and is_valid_venue(venue)):
                continue
            if title not in papers:
                citation = get_citation_from_semantic(entry)
                papers[title] = {
                    "title": title,
                    "year": year,
                    "venue": venue,
                    "citation": citation,
                    "source": "SemanticScholar",
                    "topic": topic
                }
    print(f"Topic group '{topic}': {len(papers)} papers after filtering.\n")
    return papers

# --- Plotting Functions ---
def plot_year_distribution(df, topic, filename, color):
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    df_topic = df[df['year'] >= YEAR_THRESHOLD]
    if df_topic.empty:
        print(f"No data to plot for topic {topic}.")
        return
    year_counts = df_topic['year'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    plt.bar(year_counts.index.astype(int).astype(str), year_counts.values, color=color)
    plt.xlabel("Year")
    plt.ylabel("Number of Papers")
    plt.title(f"Papers per Year (>=1980) for {topic}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved chart for '{topic}' as {filename}")
    plt.show()

def plot_summary_chart(topic_counts, filename):
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]
    colors = ["blue", "green", "purple", "crimson"]
    plt.figure(figsize=(10, 6))
    plt.bar(topics, counts, color=colors[:len(topics)])
    plt.xlabel("Topic Group")
    plt.ylabel("Total Number of Papers")
    plt.title("Total Papers per Topic Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved summary chart as {filename}")
    plt.show()

# --- Main Execution ---
def main():
    all_papers = {}
    topic_counts = {}
    color_map = {
        "RL+NS": "blue",
        "RL+MO": "green",
        "NS+MO": "purple",
        "RL+NS+MO": "crimson"
    }

    for topic, queries in combination_queries.items():
        papers = process_topic_group_semantic(topic, queries)
        all_papers.update(papers)
        topic_counts[topic] = len(papers)

        if papers:
            df_topic = pd.DataFrame(list(papers.values()))
            csv_filename = os.path.join(OUTPUT_DIR, f"{topic.replace('+','_')}_papers_semantic.csv")
            df_topic.to_csv(csv_filename, index=False)
            print(f"Saved CSV for topic group '{topic}' as {csv_filename}")

            bib_filename = os.path.join(OUTPUT_DIR, f"scraped_citations_{topic.replace('+','_')}.txt")
            with open(bib_filename, "w", encoding="utf-8") as f:
                for paper in papers.values():
                    if paper["citation"]:
                        f.write(paper["citation"] + "\n\n")
            print(f"Saved citations for topic group '{topic}' as {bib_filename}")

            plot_filename = os.path.join(OUTPUT_DIR, f"{topic.replace('+','_')}_by_year_semantic.svg")
            plot_year_distribution(df_topic, topic, plot_filename, color_map.get(topic, "black"))

    if all_papers:
        df_all = pd.DataFrame(list(all_papers.values()))
        combined_csv = os.path.join(OUTPUT_DIR, "all_topics_papers_semantic.csv")
        df_all.to_csv(combined_csv, index=False)
        print(f"Saved combined CSV for all topic groups as {combined_csv}")

    plot_summary_chart(topic_counts, os.path.join(OUTPUT_DIR, "summary_topic_counts_semantic.svg"))

if __name__ == "__main__":
    main()
```

```python
import os
import matplotlib.pyplot as plt

# Function to count citations in a file.
def count_citations(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read().strip()
    # Assume citations are separated by two newlines.
    # Filter out any empty strings.
    citations = [c for c in text.split("\n\n") if c.strip()]
    return len(citations)


### Visualization 1: Vertical Bar Chart with Number Labels ###
def plot_bar_chart(topic_counts, filename):
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]

    fig, ax = plt.subplots(figsize=(10,6))
    bars = ax.bar(topics, counts, color=["blue", "green", "purple", "crimson"])
    ax.set_xlabel("Topic Group")
    ax.set_ylabel("Total Number of Papers")
    ax.set_title("Total Papers per Topic Group")

    # Add labels on top of each bar
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom')

    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved vertical bar chart as {filename}")
    plt.show()

### Visualization 2: Pie Chart ###
def plot_pie_chart(topic_counts, filename):
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]

    fig, ax = plt.subplots(figsize=(8,8))
    wedges, texts, autotexts = ax.pie(counts, labels=topics, autopct='%1.1f%%', startangle=140,
                                      colors=["blue", "green", "purple", "crimson"])
    ax.set_title("Distribution of Papers per Topic Group")
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved pie chart as {filename}")
    plt.show()

### Visualization 3: Horizontal Bar Chart ###
def plot_horizontal_bar_chart(topic_counts, filename):
    topics = list(topic_counts.keys())
    counts = [topic_counts[t] for t in topics]

    fig, ax = plt.subplots(figsize=(10,6))
    bars = ax.barh(topics, counts, color=["blue", "green", "purple", "crimson"])
    ax.set_xlabel("Total Number of Papers")
    ax.set_ylabel("Topic Group")
    ax.set_title("Horizontal Bar Chart of Papers per Topic Group")

    # Add labels on each horizontal bar
    for bar in bars:
        width = bar.get_width()
        ax.text(width + 1, bar.get_y() + bar.get_height()/2, int(width), ha='left', va='center')

    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved horizontal bar chart as {filename}")
    plt.show()

# Define the output folder (adjust if needed)
#OUTPUT_DIR = "bipartite"

# Define file names and assign each to a topic label
files_and_topics = {
    "scraped_citations_RL_MO.txt": "RL+MO",
    "scraped_citations_RL_NS.txt": "RL+NS",
    "scraped_citations_NS_MO.txt": "NS+MO",
    "scraped_citations_RL_NS_MO.txt": "RL+NS+MO"
}

# Create a dictionary of counts per topic
topic_counts = {}
for filename, topic in files_and_topics.items():
    #full_path = os.path.join(OUTPUT_DIR, filename)
    #count = count_citations(full_path)
    count = count_citations(filename)
    topic_counts[topic] = count
    print(f"Topic '{topic}' has {count} citations.")


# Call the plotting functions and save the charts in OUTPUT_DIR.
#plot_bar_chart(topic_counts, os.path.join(OUTPUT_DIR, "bar_chart.svg"))
#plot_pie_chart(topic_counts, os.path.join(OUTPUT_DIR, "pie_chart.svg"))
#plot_horizontal_bar_chart(topic_counts, os.path.join(OUTPUT_DIR, "horizontal_bar_chart.svg"))
plot_bar_chart(topic_counts, "bar_chart.svg")
plot_pie_chart(topic_counts, "pie_chart.svg")
plot_horizontal_bar_chart(topic_counts, "horizontal_bar_chart.svg")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# --- Set working directory (adjust as needed) ---
print("Current directory:", os.getcwd())
os.chdir("/content/drive/My Drive/Colab Notebooks/Survey Paper/Neuro-Symbolic Multi-objective RL/bipartite/")
print("New directory:", os.getcwd())

# Function to load venues from a CSV file.
def load_venues_from_csv(file_path):
    df = pd.read_csv(file_path, encoding='utf-8')
    # If the CSV has a dedicated 'venue' column, use it.
    if 'venue' in df.columns:
        venues = df['venue'].dropna().astype(str).tolist()
    else:
        # Otherwise, assume the citation text is in the first column
        first_col = df.columns[0]
        citations = df[first_col].astype(str).tolist()
        venues = []
        for citation in citations:
            v = extract_venue(citation)
            if v:
                venues.append(v)
    return venues

# Function to extract the venue from a citation block.
def extract_venue(citation):
    for line in citation.splitlines():
        # Use case-insensitive matching and split on the first colon.
        if line.lower().startswith("venue:"):
            return line.split(":", 1)[1].strip()
    return None

# Function to count how many venues.
def get_venue_counts_from_list(venues):
    counts = {}
    for venue in venues:
        counts[venue] = counts.get(venue, 0) + 1
    return counts

# Save the venue counts as CSV.
def save_venue_counts(counts, filename):
    df = pd.DataFrame(list(counts.items()), columns=['Venue', 'Count'])
    df = df.sort_values(by='Count', ascending=False)
    df.to_csv(filename, index=False)
    print(f"Saved venue counts to {filename}")
    return df

# Plot a vertical bar chart (top N venues).
def plot_top_venues_bar(df, top_n=50, filename="top_venues_bar.svg"):
    top_df = df.head(top_n)
    plt.figure(figsize=(16, 10))
    bars = plt.bar(top_df['Venue'], top_df['Count'], color='skyblue')
    plt.xlabel("Venue", fontsize=10)
    plt.ylabel("Paper Count", fontsize=10)
    plt.title(f"Top {top_n} Venues by Paper Count", fontsize=12)
    plt.xticks(rotation=90, ha='right', fontsize=8)
    # Label each bar with its count.
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height, int(height),
                 ha='center', va='bottom', fontsize=8)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved vertical bar chart to {filename}")
    plt.show()

# Plot a pie chart for the top N venues.
def plot_top_venues_pie(df, top_n=50, filename="top_venues_pie.svg"):
    top_df = df.head(top_n)
    plt.figure(figsize=(10,10))
    wedges, texts, autotexts = plt.pie(top_df['Count'], labels=top_df['Venue'],
                                       autopct='%1.1f%%', startangle=140,
                                       textprops={'fontsize': 6})
    plt.title(f"Top {top_n} Venues by Paper Count (Percentage)", fontsize=10)
    plt.tight_layout()
    plt.savefig(filename, format="svg")
    print(f"Saved pie chart to {filename}")
    plt.show()

# List of CSV citation files.
citation_files = [
    "RL_NS_papers_semantic.csv",
    "RL_MO_papers_semantic.csv",
    "NS_MO_papers_semantic.csv",
    "RL_NS_MO_papers_semantic.csv"
]

# Load and concatenate venues from all CSV files.
all_venues = []
for file in citation_files:
    try:
        venues = load_venues_from_csv(file)
        print(f"Loaded {len(venues)} venues from {file}")
        all_venues.extend(venues)
    except Exception as e:
        print(f"Error loading {file}: {e}")

print(f"Total venues loaded: {len(all_venues)}")

# Generate venue counts and save as CSV.
venue_counts = get_venue_counts_from_list(all_venues)
df_counts = save_venue_counts(venue_counts, "venue_counts.csv")

# Create visualizations if data exists.
if not df_counts.empty:
    plot_top_venues_bar(df_counts, top_n=50, filename="top_venues_bar.svg")
    plot_top_venues_pie(df_counts, top_n=50, filename="top_venues_pie.svg")
else:
    print("No venue counts available to plot.")
```
